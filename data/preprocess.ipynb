```python
# Path to the original LoveDA dataset
SOURCE_DIR = "./2021LoveDA"
# Path to the processed output
OUTPUT_DIR = "./2021LoveDA_urban224"

import numpy as np
from PIL import Image
from tqdm import tqdm
```

```python
j = 0
for i in tqdm(range(1366, 2522)):
    img = np.array(Image.open(f"{SOURCE_DIR}/Train/Urban/images_png/{i}.png")).transpose(2, 0, 1)
    mask = np.array(Image.open(f"{SOURCE_DIR}/Train/Urban/masks_png/{i}.png"))
    np.save(f"{OUTPUT_DIR}/train/images/{j}.npy", img[:, :224, :224])
    np.save(f"{OUTPUT_DIR}/train/masks/{j}.npy", mask[:224, :224])
    j += 1
    np.save(f"{OUTPUT_DIR}/train/images/{j}.npy", img[:, -224:, :224])
    np.save(f"{OUTPUT_DIR}/train/masks/{j}.npy", mask[-224:, :224])
    j += 1
    np.save(f"{OUTPUT_DIR}/train/images/{j}.npy", img[:, -224:, -224:])
    np.save(f"{OUTPUT_DIR}/train/masks/{j}.npy", mask[-224:, -224:])
    j += 1
    np.save(f"{OUTPUT_DIR}/train/images/{j}.npy", img[:, :224, -224:])
    np.save(f"{OUTPUT_DIR}/train/masks/{j}.npy", mask[:224, -224:])
    j += 1
```
